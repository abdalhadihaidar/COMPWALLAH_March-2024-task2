```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

def label_encoding(df):
    categorical_columns = df.select_dtypes(include='object').columns
    label_encoder = LabelEncoder()
    df[categorical_columns] = df[categorical_columns].apply(lambda col: label_encoder.fit_transform(col))

label_encoding(data)
```

```python
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Define a function to calculate the distance using the Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = 6371 * c  # Radius of the Earth in kilometers

    return distance
```

```python

# Calculate the distance for each entry in the dataset
data['Distance (km)'] = data.apply(lambda row: calculate_distance(row['Restaurant_latitude'], row['Restaurant_longitude'], row['Delivery_location_latitude'], row['Delivery_location_longitude']), axis=1)

# Print the first few rows of the dataset with the distance column added
print(data.head())
```

```python


# vorrelations visulization
corr = data.corr()

plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True, vmin=-1.0, cmap='mako')
plt.show()
```

```python
# splitting
y = data['Time_taken(min)']
X = data.drop(['Time_taken(min)'], axis=1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error



# Define the parameter distributions for the RandomForestRegressor
param_dist = {
    'random_forest__n_estimators': randint(100, 1000),
    'random_forest__max_depth': randint(3, 10),
    # Add other hyperparameters for RandomForestRegressor here
}
# Define the pipeline with StandardScaler and RandomForestRegressor
pipe = Pipeline([('scaler', StandardScaler()), ('random_forest', RandomForestRegressor())])
# Create a randomized search object
random_search = RandomizedSearchCV(pipe, param_distributions=param_dist, n_iter=100, cv=5, scoring='neg_mean_absolute_error', n_jobs=-1, random_state=42)

# Fit the randomized search to the data
random_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = random_search.best_params_
print("Best Hyperparameters:", best_params)

# Get the best model
best_model = random_search.best_estimator_

# Evaluate the best model
y_pred = best_model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (Best Model):", mae)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

# Define the pipeline with StandardScaler and XGBRegressor
pipe = Pipeline([
    ('scaler', StandardScaler()),  # StandardScaler for preprocessing
    ('xgb', XGBRegressor(n_estimators=20, max_depth=9))  # XGBRegressor for modeling
])

# Fit the pipeline on the training data
pipe.fit(X_train, y_train)

# Evaluate on the testing data
y_pred = pipe.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```
